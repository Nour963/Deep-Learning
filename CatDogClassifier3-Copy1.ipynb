```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

DATA_DIR = '/home/nour/Documents/PetImages/'
CATEGORIES = ['Dog','Cat']


for category in CATEGORIES:
    path = os.path.join(DATA_DIR, category) #path for cat and dog directories
    for img in os.listdir(path): #for every element in the cat/dog folder
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE) #converting from bgr to gray, 
        #since in this type of classification, colors don't matter.
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break        
```

```python
#reshape images
IMG_SIZE=150
new_array = cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap="gray")
plt.show()
```

```python
from tqdm import tqdm
training_data = []
def create_train_data():
    for category in CATEGORIES:
        path = os.path.join(DATA_DIR, category) #path for cat and dog directories
        class_num = CATEGORIES.index(category)
        for img in tqdm(os.listdir(path)): #for every element in the cat/dog folder
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE) #converting from bgr to gray, 
                new_array = cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
            
create_train_data()
print(len(training_data))
                
```

```python
import random
random.shuffle(training_data) 

#print labels, 0 for dog 1 for cat
for sample in training_data:
    print(sample[1])
```

```python
X = []
y = []
for features,label in training_data:
    X.append(features)
    y.append(label)
#features HAVE to be an array, y can stay a list
X = np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,1) #1 is for pixel of gray scale, 3 if rgb
# -1 meaning whatever number, we don't know how many, but include all , and here it's number of images
```

```python
#save our X,Y to don't go over the steps again, the important is the neural network, not cleaning the data
import pickle

pickle_out = open("X.pickle","wb")
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out = open("y.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()

#for loading
#pickle_in = open("X.pickle","rb")
#X = pickle.load(pickle_in)
```

```python
import pickle
#for loading
pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)
```

```python
import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,Activation,MaxPooling2D
```

```python
X = X / 255.0 #normalization
import tensorflow as tf
from keras.callbacks import TensorBoard,ModelCheckpoint
import time


name='CatDogClassifier3-{}'.format(int(time.time()))
filepath='/home/nour/Documents/weightsCatDog.best.hdf5'

tensorboard = TensorBoard(log_dir="/home/nour/Documents/logs/{}".format(name))
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', 
                             verbose=1, save_best_only=True, mode='max')
callbacks = [tensorboard,checkpoint]

model = Sequential()
model.add(Conv2D(64,(3,3),input_shape=X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(2,2))

model.add(Flatten())


model.add(Dense(1,activation='sigmoid')) # 1 neuron, yes or no dog

model.compile(loss="binary_crossentropy",optimizer="adam", metrics=['accuracy'])
print(model.summary())
model.fit(X,y,batch_size=32,epochs=9,validation_split=0.1,callbacks=callbacks)

model.save

model_json = model.to_json()
with open("model2.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model2.h5")
print("Saved model to disk")
```
